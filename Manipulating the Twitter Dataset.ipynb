```python
import pandas as pd
import numpy as np
import re
data = pd.read_csv("C:/Users/Surya/Downloads/tweets1.csv")
data
```

```python
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)
    return input_txt
print(data)
```

```python
data['new'] = np.vectorize(remove_pattern)(data['text'],"@[\w]*")
data
```

```python
data['new'] = data['new'].str.replace("[^a-zA-Z#]"," ")
data
```

```python
data['new'] = data['new'].apply(lambda x: ' '.join([w for w in x.split() if len(w) > 3]))
data
```

```python
tokenised_tweet = data['new'].apply(lambda x: x.split())
tokenised_tweet.head()
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
tokenised_tweet = tokenised_tweet.apply(lambda x: [stemmer.stem(i) for i in x])
tokenised_tweet.head()
```
